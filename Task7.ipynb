# While Loop, else, break, continue

```python
# Shopping Cart System

Total_bill_amount =0
i=0
print("Enter the item:")

while True:
  price= int(input())
  if price <= 0:
    break
  Total_bill_amount += price
  i += 1

print("Total_bill_amount:", Total_bill_amount)

if Total_bill_amount >= 1000:
  print("You got 20% discount")
  print("You are Discount amt:",(Total_bill_amount * 0.20))
  Total_bill_amount = Total_bill_amount- (Total_bill_amount * 0.20)
  print("Net_amount:", Total_bill_amount)

elif Total_bill_amount >= 500:
  print("You got 10% discount")
  print("You are Discount amt:",(Total_bill_amount * 0.10))
  Total_bill_amount = Total_bill_amount -(Total_bill_amount * 0.10)
  print("Net_amount:", Total_bill_amount)

elif Total_bill_amount >= 200:
  print("You got 5% discount")
  print("You are Discount amt:",(Total_bill_amount * 0.05))
  Total_bill_amount = Total_bill_amount-(Total_bill_amount * 0.05)
  print("Net_amount:", Total_bill_amount)

else:
  print("  No discount")





```

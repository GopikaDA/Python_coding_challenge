```python
transaction_count = 0     # global variable


class BankAccount:

    def init(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        global transaction_count   # using global variable

        # local variable
        new_balance = self.balance + amount

        self.balance = new_balance
        transaction_count += 1

        print(f"Deposited ₹{amount} successfully.")

    def withdraw(self, amount):
        global transaction_count   # using global variable

        # local variable
        remaining_balance = self.balance - amount

        # prevent negative balance
        if remaining_balance < 0:
            print("Withdrawal failed! Insufficient balance.")
        else:
            self.balance = remaining_balance
            transaction_count += 1
            print(f"Withdrawn ₹{amount} successfully.")

    def display_balance(self):
        print(f"Account Holder : {self.account_holder}")
        print(f"Current Balance: ₹{self.balance}")

        print("Total transactions:", transaction_count)

acc1 = BankAccount("Gopika", 1000)


acc1.deposit(500)
acc1.withdraw(200)
acc1.display_balance()
```

```python
class BankAccount:

    # class variable (shared by all objects)
    transaction_count = 0

    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):

        # local variable
        new_balance = self.balance + amount

        self.balance = new_balance

        BankAccount.transaction_count += 1
        print("Deposited:", amount)

    def withdraw(self, amount):

        # local variable
        new_balance = self.balance - amount

        if new_balance < 0:
            print("Withdrawal denied. Insufficient balance.")
        else:
            self.balance = new_balance
            BankAccount.transaction_count += 1
            print("Withdrawn:", amount)

    def display_balance(self):
        print("Account Holder:", self.account_holder)
        print("Current Balance:", self.balance)


# -----------------------
# Object creation
# -----------------------
acc1 = BankAccount("Meera", 3000)

# -----------------------
# Positional argument
# -----------------------
acc1.deposit(500)

# -----------------------
# Keyword argument
# -----------------------
acc1.withdraw(amount=1000)

# -----------------------
# Negative balance test
# -----------------------
acc1.withdraw(5000)

# -----------------------
# Display
# -----------------------
acc1.display_balance()

print("Total transactions:", BankAccount.transaction_count)
```

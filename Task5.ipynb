#  Dictionary, Set, Frozenset


```python
student_record ={"raja":"B grade","Raju":"A grade","nila":"A grade"}

print(type(student_record))
print(student_record)

student_record={}
for i in range(3):
  name=input("enter the name:")
  grade=input("enter the grade:")
  student_record.update({name:grade})
print(student_record)
```

## Set and Frozenset

```python
set_empty=set()
type(set_empty)

for i in range(3):
  hobbies = input("enter the hobbies:")
  set_empty.add(hobbies)
print(set_empty)
```

```python
set_empty
frozen_set=frozenset(set_empty)
print(frozen_set)
type(frozen_set)
```

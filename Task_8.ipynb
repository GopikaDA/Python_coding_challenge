# range, for loop, enumerate, list comprehension



```python
import random
scores= [random.randint(0,100) for i in range(10)]
print(scores)
for index, score in enumerate(scores,start=1):
  print(f"The Marks are {index} = {score}")
above_average= []
for score in scores:
   if score > 75:
     above_average.append(score)
     total = sum(above_average)
     average= total/len(above_average)
print("marks Are above 75:",above_average)
print("The total mark is:",total)
print("the Average is :",average)
```

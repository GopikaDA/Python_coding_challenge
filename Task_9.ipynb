#  Built-in Functions and User-Defined Functions

```python
# Simple Calculator App
def calculator(num1,num2):
  return num1+num2,num1-num2,num1*num2,num1/num2
add,sub,mul,div=calculator(10,20)
print(f"the addition={add},the subtraction={sub},the multiplication={mul},the division={div}")
result=calculator(10,20)
print(result)
```

```python
def add(num1,num2):
  return round(num1+num2)
def sub(num1,num2):
  return round(abs(num1 - num2))
def muliply(num1,num2):
  return round(num1 * num2)
def divide(num1,num2):
  return round(num1 / num2)
a=int(input("enter the first number:"))
b=int(input("enter the second number:"))
operation = input("enter the operation like (+,-,*,/,power)")

if operation == "+":
  result=add(a,b)
  print(f"the addition of {a} and {b} is {result}")
elif operation == "-":
  result=sub(a,b)
  print(f"the subtraction of {a} and {b} is {result}")
elif operation == "*":
  result=muliply(a,b)
  print(f"the multiplication of {a} and {b} is {result}")
elif operation == "/":
  result=divide(a,b)
elif operation == "power":
  result=a**b
  print(f"{a} to the power of {b} is {result}")
else:
  print("invalid operation")








```

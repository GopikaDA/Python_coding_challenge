## Student Management System

```python
student_info =[]

def add_student():
  for i in range(3):
     name=input("Enter student name: ")
     age=int(input("Enter student age: "))
     grade=input("Enter student grade: ")
     student={'name':name,'age':age,'grade':grade}
     student_info.append(student)
     print("Student added successfully!")
     print(student_info)


add_student()
```

```python
student_info
```

```python
def search_student():
  name=input("Enter student name: ")
  for student in student_info:
    if student['name']==name:
      print("Student found!")
      print(student)
    else :
      print("Student not found!")
search_student()
```

```python
if 'student_info' not in globals():
    student_info = []
    print("Warning: 'student_info' was not found in global scope. Initializing as empty list. Please run preceding cells to load student data.")

def update_student():
  name=input("Enter student name: ")
  found = False
  for student in student_info:
    if student['name']==name:
      print("Student found!")
      print(student)
      student['age']=int(input("Enter new age: "))
      student['grade']=input("Enter new grade: ")
      print("Student updated successfully!")
      print(student)
      found = True
      break # Exit loop once student is found and updated
  if not found:
    print("Student not found!")

update_student()
```

```python
student_info
```

```python
def delete_student():
  name=input("Enter student name: ")
  found = False
  student_to_remove = None

  for student in student_info:
    if student['name'] == name:
      print("Student found!")
      print(student)
      student_to_remove = student
      found = True
      break # Exit loop once student is found

  if found:
    student_info.remove(student_to_remove)
    print("Student deleted successfully!")
    print(student_info)
  else:
    print("Student not found!")

delete_student()
```
